```python
## Summary statistcs of values at the SeriesDescription and GeoAreaCode
df.groupby(["SeriesDescription", "GeoAreaCode"])['Value'].describe()
```

```python
df["Value"] = df.groupby(["SeriesDescription", "GeoAreaCode"]).transform(lambda x: x.fillna(x.mean()))
```

```python
## Summary Statistics of BasePeriod
df.groupby(["SeriesDescription", "GeoAreaCode"])['BasePeriod'].describe()
```

```python
df["BasePeriod"] = df.groupby(["SeriesDescription", "GeoAreaCode"]).transform(lambda x: x.fillna(x.interpolate()))
```

```python
print(df["Value"].isnull().any())
print(df["BasePeriod"].isnull().any())
```

```python
cols = ["Age", "BasePeriod"]
df[cols] = df[cols].apply(pd.to_numeric, errors = "coerce", axis = 1)
```

```python
## Summary Statistics for Age
df.groupby(["SeriesDescription", "GeoAreaCode"])['Age'].describe()
```

```python
df["Age"] = df.groupby(["SeriesDescription", "GeoAreaCode"]).transform(lambda x: x.fillna(x.mean()))
```

```python
df["Education level"] = df.groupby(["SeriesDescription", "GeoAreaCode"]).transform(lambda x: x.fillna(x.mode()))
```

```python
df["Location"] = df.groupby(["SeriesDescription", "GeoAreaCode"]).transform(lambda x: x.fillna(x.mode()))
```

```python
df["Quantile"] = df.groupby(["SeriesDescription", "GeoAreaCode"]).transform(lambda x: x.fillna(x.mode()))
```

```python
df["Sex"] = df.groupby(["SeriesDescription", "GeoAreaCode"]).transform(lambda x: x.fillna(x.mode()))
```
